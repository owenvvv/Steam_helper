# Now let's start to build model!!!

## Process Description and Review Dataset

```python
import collections
SteamDocument = collections.namedtuple('SteamDocument','words appid tags split')
```

```python
%%time
steam_descrip['descrip'].astype(str)
Steam_des = []
for i in range(len(steam_descrip)):
    tokens = gensim.utils.simple_preprocess(steam_descrip.iloc[i][1])
    Steam_des.append(SteamDocument(tokens, steam_descrip.iloc[i][0],[i], 'des'))
print(len(Steam_des))
print(Steam_des[1])
```

```python
%%time
Steam_rev = []
index = 0
for i in range(len(review_postext)):
    tokens = gensim.utils.simple_preprocess(review_postext.iloc[i][1])
    if len(tokens)<57:#median word number of review
        continue
    Steam_rev.append(SteamDocument(tokens, review_postext.iloc[i][0], [index],'rev'))
    index+=1
len(Steam_rev)
```

## Put description and review together

```python
alldocs = Steam_des + Steam_rev
len(alldocs)
```

```python
print('%d docs: %d game descriptions, %d game reviews' % (len(alldocs),len(Steam_des),len(Steam_rev)))
```

## Define doc2vec Model

```python
import multiprocessing
from collections import OrderedDict

import gensim.models.doc2vec
assert gensim.models.doc2vec.FAST_VERSION > -1, "This will be painfully slow otherwise"

from gensim.models.doc2vec import Doc2Vec
```

```python
common_kwargs = dict(vector_size = 100, epochs = 40, min_count = 2,
                    sample = 0, workers = multiprocessing.cpu_count(),
                    negative = 5, hs = 0,)

simple_models = [
    #PV-DBOW
    Doc2Vec(dm=0, **common_kwargs),
    #PV-DM /average
    Doc2Vec(dm=1, window=10, alpha=0.05, comment='alpha=0.05', **common_kwargs),
    #PV-DM /concatenation
    Doc2Vec(dm=1, dm_concat = 1, window = 5, **common_kwargs),
]

for model in simple_models:
    model.build_vocab(Steam_des)
    print("%s vocabulary built and state initialized" % model)
    
```

```python
models_by_name = OrderedDict((str(model),model) for model in simple_models)
```

### Combine two models

```python
from gensim.test.test_doc2vec import ConcatenatedDoc2Vec
models_by_name['dbow+dm'] = ConcatenatedDoc2Vec([simple_models[0], simple_models[1]])
```

## Define Evaluation Methods

```python
def accuracy_for_model(model,trainset, testset, topn):
    true = 0
    all_data = 0
    for doc_id in tqdm(range(len(testset))):
        inferred_vector = model.infer_vector(testset[doc_id].words)
        sims = model.docvecs.most_similar([inferred_vector],topn = topn)
        rank = [trainset[docid].appid for docid, sim in sims]
        true += int(testset[doc_id].appid) in rank
        all_data += 1

    print('%s Accuracy: %.4f' % (model, (true/all_data)))
    
```

## Training & Per-Model Evaluation

```python
%%time
from tqdm import tqdm
for model in simple_models:
    print("Training %s" % model)
    model.train(Steam_des, total_examples = model.corpus_count, epochs = model.epochs)
    print("\nEvaluating %s" % model)
    accuracy_for_model(model = model, trainset = Steam_des, testset = Steam_rev, topn = 10)
    print('='*60)
```
