```python
import pandas as pd
import nltk
import numpy as np
from nltk.corpus import stopwords
import spacy
import collections
from tqdm import trange
from nltk.stem import WordNetLemmatizer  
from vaderSentiment import vaderSentiment
```

```python
review_small=review.sample(5000)
review_small=review_small.reset_index()
review_small=review_small[['product_id','recommended','text']]
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
end=[]
stopwordslist=set(stopwords.words('english'))
def lemmatize_all(sentence):
    wnl = WordNetLemmatizer()    
    for word, tag in nltk.pos_tag(nltk.word_tokenize(sentence)):
        if word not in stopwordslist:
            if tag.startswith('NN'):
                yield wnl.lemmatize(word, pos='n')
            elif tag.startswith('VB'):
                yield wnl.lemmatize(word, pos='v')
            elif tag.startswith('JJ'):
                yield wnl.lemmatize(word, pos='a')
            elif tag.startswith('R'):
                yield wnl.lemmatize(word, pos='r')
            else:
                yield word

for i in tqdm.trange(len(review_small)):
    text_list=review_small.iloc[i]['text']
    sentences = nltk.sent_tokenize(text_list) 
    for sentenc in sentences:
        end=end+list(lemmatize_all(sentenc))
d=collections.Counter(end)
```

```python
review=pd.read_csv('review_pre.csv')
```

```python
full_sentences=[]
for i in trange(len(review)):
    text_list=review.iloc[i]['text']
    sentences = nltk.sent_tokenize(text_list) 
    full_sentences.append(sentences)
```

```python
keywords=pd.read_excel(r'E:\steam data\Steam_Helper\data\keywords.xlsx')
keywords_class={}
keywords_class['Gameplay']=list(keywords[keywords['Gameplay'].isnull()==False]['Gameplay'])
keywords_class['Market']=list(keywords[keywords['Market'].isnull()==False]['Market'])
keywords_class['Narrative']=list(keywords[keywords['Narrative'].isnull()==False]['Narrative'])
keywords_class['Social']=list(keywords[keywords['Social'].isnull()==False]['Social'])
keywords_class['Graphics']=list(keywords[keywords['Graphics'].isnull()==False]['Graphics'])
keywords_class['Technical']=list(keywords[keywords['Technical'].isnull()==False]['Technical'])
keywords_class['Audio']=list(keywords[keywords['Audio'].isnull()==False]['Audio'])
keywords_class['Content']=list(keywords[keywords['Content'].isnull()==False]['Content'])
```

```python
all_keywords=[]
for item in keywords_class.values():
    all_keywords+=item
```

```python
all_keywords
```

```python
full_sentences_keywords=[]
for i in trange(len(full_sentences)):    
    sentence_keywords=[]
    for sentence in full_sentences[i]:
        for word in all_keywords:
            if word in sentence.split(' '):
                sentence_keywords.append(sentence)
                break
    full_sentences_keywords.append(sentence_keywords)
```

```python
review['SentencewithKeywords']=full_sentences_keywords
review
```

```python
analyzer = vaderSentiment.SentimentIntensityAnalyzer()
analyzer.polarity_scores('i hate this game')
```

```python
whichkeys=[]
for i in trange(len(review)):
    whichkey=[]
    for sentence in review.iloc[i]['SentencewithKeywords']:
        for key in keywords_class.keys():
            for word in keywords_class[key]:
                if word in sentence.split(' '):
                    a=(key,analyzer.polarity_scores(sentence)['compound'])
                    whichkey.append(a)                    
    whichkeys.append(whichkey)
```

```python
review['whichkeys']=whichkeys
review
```

```python
review['whichkeys']=whichkeys
answer=[]
ids=review['product_id'].unique()
grouped =review.groupby('product_id')
for i in trange(len(ids)):
    tmp=grouped.get_group(ids[i])
    ikey={}
    Social_score,Gameplay_score,Market_score,Narrative_score,Graphics_score,Technical_score, Audio_score, Content_score=[],[],[],[],[],[],[],[]
    for j in range(len(tmp)):
        for s in tmp.iloc[j]['whichkeys']:
            if s[0]=='Gameplay':
                Gameplay_score.append(s[1])
            elif s[0]=='Social':
                Social_score.append(s[1])
            elif s[0]=='Market':
                Market_score.append(s[1])
            elif s[0]=='Narrative':
                Narrative_score.append(s[1])
            elif s[0]=='Graphics':
                Graphics_score.append(s[1])
            elif s[0]=='Technical':
                Technical_score.append(s[1])
            elif s[0]=='Audio':
                Audio_score.append(s[1])
            elif s[0]=='Content':
                Content_score.append(s[1])      
    ikey['Gameplay']=np.mean(Gameplay_score)
    ikey['Social']=np.mean(Social_score)
    ikey['Market']=np.mean(Market_score)    
    ikey['Narrative']=np.mean(Narrative_score)
    ikey['Graphics']=np.mean(Graphics_score)
    ikey['Technical']=np.mean(Technical_score)
    ikey['Audio']=np.mean(Audio_score)
    ikey['Content']=np.mean(Content_score)
    answer.append(ikey)
```

```python
answerpd = pd.DataFrame(answer, index = ids)
answerpd.sort_values('Audio',ascending = False)
```
